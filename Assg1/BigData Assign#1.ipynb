```python
#loading and procesing the data
import matplotlib
import scipy
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston

boston = datasets.load_boston()
dataset = load_boston()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['MEDV'] = dataset.target
prices = df['MEDV']
df3 = pd.DataFrame(boston.target,columns=["price"])

data1 = pd.DataFrame(data= np.c_[boston['data'],],
                     columns= boston['feature_names'])

subset = pd.DataFrame(data1, columns = ['NOX', 'AGE', 'DIS'])
subset['price'] = df3.iloc[:,0].values
#print(subset)
```

```python
#statistics
subset.describe()
```

```python
#histograms
subset['AGE'].plot(kind = 'hist', bins = 50)
plt.title('AGE Features')
plt.xlabel('AGE Groups')
plt.show()
```

```python
#histograms
subset['NOX'].plot(kind = 'hist', bins = 50)
plt.title('NOX Features')
plt.xlabel('Nitrous Levels')
plt.show()
```

```python
#histograms
subset['DIS'].plot(kind = 'hist', bins = 50)
plt.title('Distance Features')
plt.xlabel('Distances')
plt.show()
```

```python
#scatter plots and its pearson coefficent
subset.plot(kind = 'scatter', x = 'AGE' , y = 'price')
pearsonValue = subset['AGE'].corr(subset['price'])
plt.title('AGE Vs PRICE')
plt.text(110, 40, 'The Pearson coefficient of AGE vs Price is ' + str(pearsonValue), fontsize=14)
plt.show()
```

```python
#scatter plots
subset.plot(kind = 'scatter', x = 'DIS' , y = 'price')
pearsonValue = subset['DIS'].corr(subset['price'])
plt.title('DISTANCE Vs PRICE')
plt.text(13, 40, 'The Pearson coefficient of Distance vs Price is ' + str(pearsonValue), fontsize=14)
plt.show()
```

```python
#scatter plots
subset.plot(kind = 'scatter', x = 'NOX' , y = 'price')
pearsonValue = subset['NOX'].corr(subset['price'])
plt.text(0.95, 40, 'The Pearson coefficient of Nitrous Levels vs Price is ' + str(pearsonValue), fontsize=14)
plt.title('Nitrous Levels Vs PRICE')
plt.show()
```

# Correlation Evaluation
1) NOX has the worst relation with price because you can see that there are many points all over the plot with same price values    and theoratically the Pearson coffietient value is the lowest 

2) Distance has the best relation with price because you can see that there are many grouped points in the plot 
   which mean they are dependent with the price values and theoratically the Pearson coffietient value is the highest

```python
a = scipy.spatial.distance.euclidean(subset['AGE'],subset['AGE'])
b = scipy.spatial.distance.euclidean(subset['AGE'],subset['DIS'])
c = scipy.spatial.distance.euclidean(subset['AGE'],subset['NOX'])
d = scipy.spatial.distance.euclidean(subset['AGE'],subset['price'])
```

```python
a1 = scipy.spatial.distance.euclidean(subset['DIS'],subset['AGE'])
b1 = scipy.spatial.distance.euclidean(subset['DIS'],subset['DIS'])
c1 = scipy.spatial.distance.euclidean(subset['DIS'],subset['NOX'])
d1 = scipy.spatial.distance.euclidean(subset['DIS'],subset['price'])
```

```python
a2 = scipy.spatial.distance.euclidean(subset['NOX'],subset['AGE'])
b2 = scipy.spatial.distance.euclidean(subset['NOX'],subset['DIS'])
c2 = scipy.spatial.distance.euclidean(subset['NOX'],subset['NOX'])
d2 = scipy.spatial.distance.euclidean(subset['NOX'],subset['price'])
```

```python
a3 = scipy.spatial.distance.euclidean(subset['price'],subset['AGE'])
b3 = scipy.spatial.distance.euclidean(subset['price'],subset['DIS'])
c3 = scipy.spatial.distance.euclidean(subset['price'],subset['NOX'])
d3 = scipy.spatial.distance.euclidean(subset['price'],subset['price'])
```
